- Plot the graph of $f(x) = 0.2\cdot x^2\cdot \sin\dfrac{1}{x}$ over $[0.001, 0.1]$. Make it dashed and green

```python
x = np.linspace(0.001,0.1,1000)
y = 0.2*(x**2)*np.sin(1/x)
z = 0.2*(x**2)
```

```python
x0 = np.array([1/(np.pi/2 + 2*np.pi*k)for k in range(2,9)])
y0 = 0.2*(x0**2)
```

```python
plt.figure(figsize=(10,8))
plt.plot(x,y, c = 'green', linestyle='dashed', label= 'plot of f(x)') #'g--'
plt.plot(x,z, 'r-', lw=2, label='plot of x^2')
plt.plot(x,-z, 'r-', lw=2, label='plot of upper')
plt.scatter(x0,y0, c='blue', marker="x", s = 60)
plt.title('Graph of ' + r'$\sigma{x} = 0.2\cdot x^2\cdot \sin\dfrac{1}{x}$'  )
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
```
