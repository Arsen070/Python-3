# Matplotlib

- Make a figure and add some Geometric shapes to the axis (Rectangle, Circle, Polygon) by using `add_patch`
- Use transparency alpha

```python
fig = plt.figure()

rect = plt.Rectangle((1,1),height= 1, width= 2, color='r')
circ = plt.Circle(xy=(2,2), radius=1, alpha= 0.4)
poly = plt.Polygon([[0,0],[1,0],[1,1]])

ax = fig.add_subplot(1,1,1)
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(poly)
ax.set_xlim([0,5])
ax.set_ylim([0,4])
ax.axis('equal')
```

- Use the `seaborn`-s `displot` for making different visualizations on one of the numerical/categorical variables of the penguins dataset
    * Change the kind to "hist", "kde"
    * Use bins, binwidth with Histograms
    * Use `stat="density"` for Histogram
    * Use "fill" with "kde"
    * Use `bw_adjust` for `kde`
    * Add datapoints using the rug
    * Use separate colors for each Species; use `multiple = 'stack'` with this
    * Use `col` to separate over the "sex" variable

```python
peng.columns
```

```python
sns.displot(x = 'flipper_length_mm', data = peng,kind= 'kde', fill = True)
```

```python
sns.displot(x = 'flipper_length_mm', data = peng, stat = 'density') 
sns.displot(x = 'flipper_length_mm', data = peng, kind = 'kde')
```

```python
sns.displot(x = 'flipper_length_mm', data = peng,kind= 'kde', hue = 'species', col = 'sex', fill = True)
```

```python
sns.displot(x = 'flipper_length_mm', data = peng,kind= 'kde', hue = 'sex', fill = True, multiple = 'stack')
```

- Use `displot` for 2 numerical variables of penguins dataset;
    * Separate over Species, over Sex

```python
sns.displot(x= 'flipper_length_mm', y = 'body_mass_g', data = peng, kind = 'kde')
```

- Use `jointplot` for 2 numerical variables of penguins dataset;
    * Separate over Species, over Sex
    * Use `kde`

```python
sns.jointplot(x= 'flipper_length_mm', y = 'body_mass_g', data = peng, kind = 'kde', hue = 'sex')
```

- Do `pairplot` for penguins dataset

```python
sns.pairplot(peng, hue = 'sex')
```

- Use `relplot` (Scatterplot in `seaborn`) for ploting one numerical variable versus the other in the penguins dataset
    * Use `size`, `hue`, `col`, `style` to separate different subclasses

```python
sns.relplot(x= 'flipper_length_mm', y = 'body_mass_g', data = peng, hue = 'sex', size = 'species')
```

```python
sns.relplot(x= 'flipper_length_mm', y = 'body_mass_g', data = peng, hue = 'sex', style = 'species', col = 'island')
```
